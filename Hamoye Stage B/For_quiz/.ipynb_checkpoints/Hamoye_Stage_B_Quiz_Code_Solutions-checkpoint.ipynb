```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
energy_data = pd.read_csv('energydata_complete.csv')
```

```python
energy_data.head()
```

```python
energy_data.isna().sum()
```

## Question 12 solutions

```python
lin_mod= LinearRegression()
```

```python
first_x = energy_data[['T2']]
first_y = energy_data[['T6']]
```

```python
lin_mod.fit(first_x, first_y)
```

```python
prd_lin_mod = lin_mod.predict(first_x)
```

```python
# question 12
round((r2_score(first_y, prd_lin_mod)), 2)
```

## Question 13 - 17 solutions

```python
X= energy_data.drop(['date', 'lights'], axis= 1)
```

```python
normalized_data = MinMaxScaler()
```

```python
X = pd.DataFrame(normalized_data.fit_transform(X.values), columns=X.columns, index=X.index)
```

```python
X
```

```python
new_x = X.drop('Appliances', axis= 1)
new_y = X[['Appliances']]
```

```python
new_x
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_x, new_y, test_size=0.30, random_state=42)
```

```python
new_lin_mod = LinearRegression()
```

```python
new_lin_mod.fit(X_train, y_train)
```

```python
new_pred = new_lin_mod.predict(X_test)
```

```python
# question 13
round(mean_absolute_error(y_test, new_pred), 2)
```

```python
# question 14
round(mean_squared_error(y_test, new_pred), 2)
```

```python
# question 15
round(np.sqrt(mean_squared_error(y_test, new_pred)), 3)
```

```python
# question 16
round(r2_score(y_test, new_pred), 2)
```

```python
# question 17
coeff = pd.Series(new_lin_mod.coef_[0], index = new_x.columns)
```

```python
coeff[coeff.values == np.min(new_lin_mod.coef_)]
```

```python
coeff[coeff.values == np.max(new_lin_mod.coef_)]
```

## Question 18 solutions

```python
ridge = Ridge(alpha= 0.4)
```

```python
ridge.fit(X_train, y_train)
```

```python
ridge_pred = ridge.predict(X_test)
```

```python
# question 18
round(np.sqrt(mean_squared_error(y_test, ridge_pred)), 3)
```

## Question 19 - 20 solutions

```python
lasso = Lasso(alpha= 0.001)
```

```python
lasso.fit(X_train, y_train)
```

```python
# question 19
sum(lasso.coef_ > 0) + sum(lasso.coef_ < 0)
```

```python
lasso_pred = lasso.predict(X_test)
```

```python
# question 20
round(np.sqrt(mean_squared_error(y_test, lasso_pred)), 3)
```
